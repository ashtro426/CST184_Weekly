```r
Learn how to fit a linear model in R, interpret its summary statistics (slope, intercept, p-value, R²),
and connect this to a real-world scenario addressing SDG 3: Good Health and Well-Being

You work with a public health research team investigating whether increased sugar intake is associated with higher obesity rates. You’ve collected data from 50 countries on:

x: Average daily sugar intake per person (in grams)
y: Obesity rate (% of adult population)
Your task is to build a linear model and interpret its meaning.
```

```r
set.seed(2025)

sugar_intake <- runif(50, min = 30, max = 120)  # grams/day
obesity_rate <- 0.35 * sugar_intake + rnorm(50, mean = 0, sd = 5)

health_df <- data.frame(sugar_intake, obesity_rate)
head(health_df)
```

```r
library(ggplot2)

ggplot(health_df, aes(x = sugar_intake, y = obesity_rate)) +
  geom_point(color = "darkgreen") +
  labs(
    title = "Sugar Intake vs. Obesity Rate",
    x = "Daily Sugar Intake (g)",
    y = "Obesity Rate (%)"
  ) +
  theme_minimal()
```

```r
model <- lm(obesity_rate ~ sugar_intake, data = health_df)
summary(model)
```

```r
ggplot(health_df, aes(x = sugar_intake, y = obesity_rate)) +
  geom_point(color = "darkgreen") +
  geom_smooth(method = "lm", se = TRUE, color = "red") +
  labs(
    title = "Regression Model: Sugar Intake vs. Obesity Rate",
    x = "Sugar Intake (grams/day)",
    y = "Obesity Rate (%)"
  ) +
  theme_minimal()
```

```r
Learn how to apply a two-sample t-test and calculate a 95% confidence interval in R using a 
simplified real-world scenario aligned with SDG 3 (Health) and SDG 13 (Climate Action).

You work for a health research agency. You want to test whether people living in high air pollution areas
have higher asthma rates than those in low pollution areas.

You collect data from two samples (30 individuals each) and run a two-sample t-test in R
```

```r
# Simulate asthma data (synthetic for learning)
set.seed(42)

low_pollution <- rnorm(30, mean = 11, sd = 3)   # Group A
high_pollution <- rnorm(30, mean = 14, sd = 4)  # Group B

# Combine into a dataframe
asthma_df <- data.frame(
  rate = c(low_pollution, high_pollution),
  group = c(rep("Low Pollution", 30), rep("High Pollution", 30))
)

head(asthma_df)
```

```r
t_result <- t.test(rate ~ group, data = asthma_df)
print(t_result)
```

```r
library(ggplot2)

ggplot(asthma_df, aes(x = group, y = rate, fill = group)) +
  geom_boxplot() +
  labs(
    title = "Asthma Rates by Pollution Exposure",
    y = "Asthma Rate (per 1000 people)",
    x = "Region"
  ) +
  theme_minimal()
```

```python
import pandas as pd
df = pd.read_csv("week6_grouping_filtering.csv")
df.head()
```

```python
df_cleaned = df.dropna(subset=['Attendance'])
```

```python
region_year = (
    df_cleaned.groupby(['Region', 'Year'])['Attendance']
    .mean()
    .reset_index()
)
print(region_year)
```

```python
gender_region = (
    df_cleaned.groupby(['Region', 'Gender'])['Attendance']
    .mean()
    .reset_index()
)
print(gender_region)
```

```python
region_avg = df_cleaned.groupby('Region')['Attendance'].mean()
print("High Attendance:\n", region_avg[region_avg > 80])
print("Low Attendance:\n", region_avg[region_avg < 60])
```

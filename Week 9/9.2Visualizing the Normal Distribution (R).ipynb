```r
install.packages("ggplot2")
library(ggplot2)
```

```r
mean_temp <- 0     # average anomaly = 0째C
sd_temp <- 1       # standard deviation = 1째C

# Create x-values for the curve
x <- seq(-4, 4, length.out = 200)

# Generate y-values using the density function
y <- dnorm(x, mean = mean_temp, sd = sd_temp)

# Create data frame
df <- data.frame(x, y)

# Plot the normal curve
ggplot(df, aes(x, y)) +
  geom_line(color = "steelblue", linewidth = 1.2) +
  labs(
    title = "Normal Distribution of Temperature Anomalies",
    x = "Temperature Anomaly (째C)",
    y = "Density"
  ) +
  theme_minimal()
```

```r
mean_temp <- 1.5
sd_temp <- 2

x <- seq(-6, 6, length.out = 200)
y <- dnorm(x, mean = mean_temp, sd = sd_temp)
df <- data.frame(x, y)

ggplot(df, aes(x, y)) +
  geom_line(color = "firebrick", size = 1.2) +
  labs(
    title = "Modified Normal Distribution (Mean = 1.5, SD = 2)",
    x = "Temperature Anomaly (째C)",
    y = "Density"
  ) +
  theme_minimal()
```

```r
# Load ggplot2
library(ggplot2)

# Parameters
n <- 10
p <- 0.7
x <- 0:n

# Binomial probabilities
y <- dbinom(x, size = n, prob = p)

# Put into a data frame for ggplot2
df <- data.frame(
  vaccinated = x,
  probability = y
)

# Expected value
expected_value <- n * p

# Plot
ggplot(df, aes(x = vaccinated, y = probability)) +
  geom_bar(stat = "identity", fill = "skyblue", color = "black") +
  geom_vline(xintercept = expected_value, color = "red", linetype = "dashed", size = 1) +
  labs(
    title = "Probability Distribution of Vaccinated People in a Group of 10",
    x = "Number of Vaccinated People",
    y = "Probability"
  ) +
  theme_minimal()
```

```r
# Load ggplot2
library(ggplot2)

# Parameters
lambda <- 4
x <- 0:15

# Poisson probabilities
y <- dpois(x, lambda = lambda)

# Put into a data frame
df <- data.frame(
  events = x,
  probability = y
)

# Plot
ggplot(df, aes(x = events, y = probability)) +
  geom_bar(stat = "identity", fill = "lightgreen", color = "black") +
  labs(
    title = "Poisson Distribution of Extreme Weather Events",
    x = "Number of Events",
    y = "Probability"
  ) +
  theme_minimal()
```

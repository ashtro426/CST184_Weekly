```r
# Step 1 – Define a Prior (Beta Distribution)
alpha_prior <- 2
beta_prior  <- 2

# Step 2 – Add Observed Data
successes <- 38
failures  <- 12

# Step 3 – Update to Posterior
alpha_post <- alpha_prior + successes  # 2 + 38 = 40
beta_post  <- beta_prior  + failures   # 2 + 12 = 14

# Step 4 – Simulate Posterior Samples
posterior <- rbeta(10000, alpha_post, beta_post)

# Step 5 – Visualize the Posterior Distribution
hist(posterior, breaks = 50, col = "skyblue",
     main = "Posterior Distribution of Household Access",
     xlab = "Proportion of Households with Access")

# Step 6 – Estimate Probability Access > 70%
mean(posterior > 0.7)
```

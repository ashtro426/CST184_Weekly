```r
supply_mean <- 45
supply_sd <- 5
demand_mean <- 40
demand_sd <- 6

results <- replicate(1000, {
  supply <- rnorm(30, mean = supply_mean, sd = supply_sd)
  demand <- rnorm(30, mean = demand_mean, sd = demand_sd)
  all(supply >= demand)
})

mean(results)
```
